```python
import os
import zipfile
import urllib.request

zip_path = "/content/arabic-speech-corpus.zip"
extract_path = "/content"
url = "https://en.arabicspeechcorpus.com/arabic-speech-corpus.zip"

# Download the dataset
print("Downloading dataset...")
urllib.request.urlretrieve(url, zip_path)
print("Download complete.")

# Unzip the dataset
print("Extracting files...")
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Extraction complete.")
```

```python
import os
import torch
import torchaudio
import numpy as np
from speechbrain.inference.speaker import EncoderClassifier
import kaldiio

# Choose device: use GPU if available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Initialize the SpeechBrain speaker classifier
classifier = EncoderClassifier.from_hparams(
    source="speechbrain/spkrec-xvect-voxceleb",
    savedir="pretrained_models/spkrec-xvect-voxceleb",
    run_opts={"device": str(device)}
)
classifier.to(device)

# Specify the folder that contains your audio files
audio_folder = "/content/arabic-speech-corpus/wav"  # <-- change to your folder
# Specify output folder to save Kaldi-style embeddings (ark and scp files)
output_folder = "train_speaker_embeddings"
os.makedirs(output_folder, exist_ok=True)
ark_path = os.path.join(output_folder, "train_spk_embed.ark")
scp_path = os.path.join(output_folder, "train_spk_embed.scp")

# Create a Kaldi writer to write in ark/scp format
with kaldiio.WriteHelper(f"ark,scp:{ark_path},{scp_path}") as writer:
    # Process every WAV file in the folder
    for filename in sorted(os.listdir(audio_folder)):
        if filename.lower().endswith(".wav"):
            # Use the filename (without extension) as the utterance ID
            utt_id = os.path.splitext(filename)[0]
            file_path = os.path.join(audio_folder, filename)

            # Load audio using torchaudio; signal shape is (channels, time)
            signal, fs = torchaudio.load(file_path)

            # If multi-channel, average channels to create a mono signal
            if signal.ndim > 1 and signal.shape[0] > 1:
                signal = torch.mean(signal, dim=0, keepdim=True)

            # Prepare the signal for batch processing: shape -> (1, time)
            signal = signal.squeeze(0).unsqueeze(0)

            # Compute the speaker embedding. The classifier returns a tensor of shape (1, emb_dim)
            emb = classifier.encode_batch(signal)  # shape: (1, embedding_dim)
            emb_np = emb.detach().cpu().numpy().squeeze()  # convert to numpy, shape: (embedding_dim,)

            # Write the embedding with the utterance ID
            writer[utt_id] = emb_np

print("Speaker embeddings have been saved to:")
print("ARK file:", ark_path)
print("SCP file:", scp_path)
```
